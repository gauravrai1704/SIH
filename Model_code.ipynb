```python
# Install required dependencies
!pip install tensorflow==2.19.0
!pip install shap streamlit -q
!pip install scikit-optimize -q
!pip install google-generativeai -q
!pip install plotly -q
!pip install kagglehub -q
!pip install pyngrok -q
!pip install spectral -q
!pip install rasterio -q
!pip install geopandas -q
!pip install earthpy -q
!pip install scikit-image -q
!pip install scipy -q
!pip install streamlit-folium -q
!pip install Pillow -q
!pip install xgboost -q
!pip install imbalanced-learn -q
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from scipy import io
import cv2
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, mean_squared_error, r2_score, mean_absolute_error
## MODIFICATION START: Import DummyRegressor for benchmarking ##
from sklearn.dummy import DummyRegressor
## MODIFICATION END ##
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, LSTM, Conv2D, MaxPooling2D, Flatten, Dropout, Input, concatenate, TimeDistributed, GlobalAveragePooling2D
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.applications import ResNet50, EfficientNetB0, MobileNetV2
from tensorflow.keras.applications.resnet50 import preprocess_input as resnet_preprocess
from tensorflow.keras.applications.efficientnet import preprocess_input as efficientnet_preprocess
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input as mobilenet_preprocess
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import rasterio
from rasterio.plot import show
import xml.etree.ElementTree as ET
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Mount Google Drive and set up paths
try:
    from google.colab import drive
    drive.mount('/content/drive', force_remount=True)
    BASE_PATH = Path('/content/drive/MyDrive')
except Exception:
    print("Google Drive not found. Using local paths.")
    BASE_PATH = Path('./')

DATASET_PATH = BASE_PATH / 'dataset'

# Check what files we have
print("Available files in dataset:")
for item in DATASET_PATH.iterdir():
    print(f"  - {item.name}")

# 1. Load and process the datasets
def load_salinas_data():
    """Load and process Salinas hyperspectral dataset"""
    salinas_path = DATASET_PATH / 'Salinas.mat'
    salinas_gt_path = DATASET_PATH / 'Salinas_gt.mat'
    salinas_corrected_path = DATASET_PATH / 'Salinas_corrected.mat'

    salinas_data = None
    salinas_gt = None

    if salinas_path.exists():
        salinas_data = io.loadmat(salinas_path)['salinas']
        print(f"Salinas data shape: {salinas_data.shape}")

    if salinas_gt_path.exists():
        salinas_gt = io.loadmat(salinas_gt_path)['salinas_gt']
        print(f"Salinas ground truth shape: {salinas_gt.shape}")

    return salinas_data, salinas_gt

def load_crop_yield_data():
    """Load crop yield and disease data"""
    csv_path = DATASET_PATH / 'crop_yield_and_disease_data.csv'

    if csv_path.exists():
        df = pd.read_csv(csv_path)
        print(f"Crop yield data shape: {df.shape}")
        print(f"Columns: {df.columns.tolist()}")
        return df
    else:
        print("Crop yield data not found")
        return None

def load_farm_insects():
    """Load farm insects image data for pest detection"""
    insects_path = DATASET_PATH / 'Farm-Insects'

    if insects_path.exists():
        # Get all image files and their labels
        image_files = []
        labels = []
        label_map = {}

        for i, class_dir in enumerate(insects_path.iterdir()):
            if class_dir.is_dir():
                label_map[i] = class_dir.name
                for img_file in class_dir.glob('*.*'):
                    if img_file.suffix.lower() in ['.jpg', '.jpeg', '.png', '.bmp']:
                        image_files.append(img_file)
                        labels.append(i)

        print(f"Found {len(image_files)} insect images across {len(label_map)} classes")
        return image_files, labels, label_map
    else:
        print("Farm-Insects directory not found")
        return [], [], {}

def load_lc08_data():
    """Load Landsat 8 data"""
    lc08_path = DATASET_PATH / 'LC08'

    if lc08_path.exists():
        # Look for Landsat 8 GeoTIFF files
        tiff_files = list(lc08_path.glob('**/*.TIF')) + list(lc08_path.glob('**/*.tif'))

        if tiff_files:
            print(f"Found {len(tiff_files)} Landsat 8 GeoTIFF files")

            # Load the first band as an example
            with rasterio.open(tiff_files[0]) as src:
                band_data = src.read(1)
                transform = src.transform
                crs = src.crs

            print(f"Band shape: {band_data.shape}, CRS: {crs}")
            return tiff_files, band_data, transform, crs
        else:
            print("No GeoTIFF files found in LC08 directory")
            return [], None, None, None
    else:
        print("LC08 directory not found")
        return [], None, None, None

def load_safe_data():
    """Load Sentinel SAFE format data with proper preprocessing"""
    safe_path = DATASET_PATH / '*.SAFE'
    safe_folders = list(DATASET_PATH.glob('*.SAFE'))

    if safe_folders:
        print(f"Found {len(safe_folders)} SAFE folders")

        # Process the first SAFE folder
        safe_folder = safe_folders[0]
        print(f"Processing SAFE folder: {safe_folder.name}")

        # Look for image data (typically in GRANULE subfolder)
        granule_path = safe_folder / 'GRANULE'
        if granule_path.exists():
            # Look for image files
            img_files = list(granule_path.glob('**/*.jp2')) + list(granule_path.glob('**/*.JP2'))
            print(f"Found {len(img_files)} image files in SAFE folder")

            if img_files:
                # Load and preprocess Sentinel-2 data (normalize reflectance)
                band_data = {}
                band_names = {
                    'B02': 'blue', 'B03': 'green', 'B04': 'red', 'B08': 'nir',
                    'B05': 'rededge1', 'B06': 'rededge2', 'B07': 'rededge3',
                    'B8A': 'rededge4', 'B11': 'swir1', 'B12': 'swir2'
                }

                for img_file in img_files:
                    for band_code, band_name in band_names.items():
                        if band_code in str(img_file):
                            with rasterio.open(img_file) as src:
                                # Normalize reflectance values (divide by 10000)
                                band_data[band_name] = src.read(1) / 10000.0
                                print(f"Loaded and normalized {band_name} band")

                # Ensure bands align spatially by checking dimensions
                first_band = list(band_data.values())[0] if band_data else None
                if first_band is not None:
                    aligned_bands = {}
                    for band_name, data in band_data.items():
                        if data.shape == first_band.shape:
                            aligned_bands[band_name] = data
                        else:
                            print(f"Band {band_name} has different dimensions, resizing...")
                            aligned_bands[band_name] = cv2.resize(data, (first_band.shape[1], first_band.shape[0]))

                    print(f"Loaded {len(aligned_bands)} aligned bands")
                    return aligned_bands, safe_folder
                else:
                    print("No bands loaded successfully")
                    return {}, safe_folder
        else:
            print("No GRANULE folder found in SAFE directory")
            return {}, None
    else:
        print("No SAFE folders found")
        return {}, None

# 2. Vegetation Indices Calculation
def calculate_ndvi(red_band, nir_band):
    """Calculate Normalized Difference Vegetation Index"""
    return (nir_band - red_band) / (nir_band + red_band + 1e-10)

def calculate_evi(blue_band, red_band, nir_band):
    """Calculate Enhanced Vegetation Index"""
    return 2.5 * (nir_band - red_band) / (nir_band + 6 * red_band - 7.5 * blue_band + 1 + 1e-10)

def calculate_ndwi(green_band, nir_band):
    """Calculate Normalized Difference Water Index"""
    return (green_band - nir_band) / (green_band + nir_band + 1e-10)

def calculate_msavi2(nir_band, red_band):
    """Calculate Modified Soil Adjusted Vegetation Index 2"""
    return (2 * nir_band + 1 - np.sqrt((2 * nir_band + 1)**2 - 8 * (nir_band - red_band))) / 2

def calculate_sentinel_indices(sentinel_bands):
    """Calculate vegetation indices from Sentinel-2 data"""
    indices = {}

    if 'red' in sentinel_bands and 'nir' in sentinel_bands:
        indices['ndvi'] = calculate_ndvi(sentinel_bands['red'], sentinel_bands['nir'])

    if 'blue' in sentinel_bands and 'red' in sentinel_bands and 'nir' in sentinel_bands:
        indices['evi'] = calculate_evi(sentinel_bands['blue'], sentinel_bands['red'], sentinel_bands['nir'])

    if 'green' in sentinel_bands and 'nir' in sentinel_bands:
        indices['ndwi'] = calculate_ndwi(sentinel_bands['green'], sentinel_bands['nir'])

    if 'red' in sentinel_bands and 'nir' in sentinel_bands:
        indices['msavi2'] = calculate_msavi2(sentinel_bands['nir'], sentinel_bands['red'])

    return indices

def extract_vegetation_indices(hyperspectral_data):
    """Extract various vegetation indices from hyperspectral data"""
    # For demonstration, we'll assume bands (approximate wavelengths)
    # These would need to be adjusted based on your actual sensor bands
    blue_band = hyperspectral_data[:, :, 10]  # ~450nm
    green_band = hyperspectral_data[:, :, 20]  # ~550nm
    red_band = hyperspectral_data[:, :, 30]    # ~650nm
    nir_band = hyperspectral_data[:, :, 40]    # ~800nm

    ndvi = calculate_ndvi(red_band, nir_band)
    evi = calculate_evi(blue_band, red_band, nir_band)
    ndwi = calculate_ndwi(green_band, nir_band)
    msavi2 = calculate_msavi2(nir_band, red_band)

    return ndvi, evi, ndwi, msavi2

def extract_landsat_indices(tiff_files):
    """Extract vegetation indices from Landsat 8 data"""
    if not tiff_files:
        return None, None, None, None

    # Find specific bands (this would need to be adjusted based on your file naming)
    band_files = {}

    for tiff_file in tiff_files:
        if 'B2' in tiff_file.name:  # Blue
            band_files['blue'] = tiff_file
        elif 'B3' in tiff_file.name:  # Green
            band_files['green'] = tiff_file
        elif 'B4' in tiff_file.name:  # Red
            band_files['red'] = tiff_file
        elif 'B5' in tiff_file.name:  # NIR
            band_files['nir'] = tiff_file

    # Load bands and calculate indices
    indices = {}
    if all(k in band_files for k in ['red', 'nir']):
        with rasterio.open(band_files['red']) as src:
            red_band = src.read(1)
        with rasterio.open(band_files['nir']) as src:
            nir_band = src.read(1)

        indices['ndvi'] = calculate_ndvi(red_band, nir_band)

        if 'blue' in band_files:
            with rasterio.open(band_files['blue']) as src:
                blue_band = src.read(1)
            indices['evi'] = calculate_evi(blue_band, red_band, nir_band)

        if 'green' in band_files:
            with rasterio.open(band_files['green']) as src:
                green_band = src.read(1)
            indices['ndwi'] = calculate_ndwi(green_band, nir_band)

    return indices

# 3. Enhanced Pest Detection Model with Transfer Learning
def build_pest_detection_model_transfer(input_shape, num_classes, model_type='efficientnet'):
    """Build a pest detection model using transfer learning"""
    # Create base model
    if model_type == 'resnet':
        base_model = ResNet50(weights='imagenet', include_top=False, input_shape=input_shape)
        preprocess_fn = resnet_preprocess
    elif model_type == 'mobilenet':
        base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=input_shape)
        preprocess_fn = mobilenet_preprocess
    else:  # efficientnet
        base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=input_shape)
        preprocess_fn = efficientnet_preprocess

    # Freeze base model layers
    base_model.trainable = False

    # Add custom head
    inputs = Input(shape=input_shape)
    x = preprocess_fn(inputs)
    x = base_model(x, training=False)
    x = GlobalAveragePooling2D()(x)
    x = Dense(128, activation='relu')(x)
    x = Dropout(0.5)(x)
    outputs = Dense(num_classes, activation='softmax')(x)

    model = Model(inputs, outputs)

    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    return model, preprocess_fn

def prepare_insect_data_augmented(image_files, labels, img_size=(224, 224), augment=True):
    """Prepare insect image data for training with augmentation"""
    images = []
    processed_labels = []

    # Create data augmentation generator
    if augment:
        datagen = ImageDataGenerator(
            rotation_range=30,
            width_shift_range=0.2,
            height_shift_range=0.2,
            shear_range=0.2,
            zoom_range=0.2,
            horizontal_flip=True,
            vertical_flip=True,
            brightness_range=[0.8, 1.2],
            fill_mode='nearest'
        )

    for i, img_file in enumerate(image_files):
        try:
            # Read and resize image
            img = cv2.imread(str(img_file))
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            img = cv2.resize(img, img_size)

            # Apply augmentation if needed
            if augment:
                img = datagen.random_transform(img)

            # Normalize pixel values
            img = img / 255.0

            images.append(img)
            processed_labels.append(labels[i])
        except Exception as e:
            print(f"Error processing image {img_file}: {e}")

    return np.array(images), np.array(processed_labels)

def calculate_class_weights(labels):
    """Calculate class weights for imbalanced datasets"""
    from sklearn.utils.class_weight import compute_class_weight
    unique_labels = np.unique(labels)
    class_weights = compute_class_weight('balanced', classes=unique_labels, y=labels)
    return dict(zip(unique_labels, class_weights))

# 4. Enhanced Temporal Analysis
def analyze_temporal_trends(crop_yield_df):
    """Analyze temporal trends in crop yield and sensor data"""
    if crop_yield_df is None:
        return None

    # Check if we have date information
    date_columns = [col for col in crop_yield_df.columns if 'date' in col.lower() or 'time' in col.lower()]

    if date_columns:
        # Use the first date column found
        date_col = date_columns[0]

        # Convert to datetime
        crop_yield_df[date_col] = pd.to_datetime(crop_yield_df[date_col])

        # Sort by date
        crop_yield_df = crop_yield_df.sort_values(date_col)

        # Plot temporal trends
        fig, axes = plt.subplots(2, 2, figsize=(15, 10))

        # Plot yield over time if available
        if 'yield' in crop_yield_df.columns:
            axes[0, 0].plot(crop_yield_df[date_col], crop_yield_df['yield'], 'o-')
            axes[0, 0].set_title('Crop Yield Over Time')
            axes[0, 0].set_xlabel('Date')
            axes[0, 0].set_ylabel('Yield')
            axes[0, 0].tick_params(axis='x', rotation=45)

        # Plot NDVI over time if available
        ndvi_cols = [col for col in crop_yield_df.columns if 'ndvi' in col.lower()]
        if ndvi_cols:
            for col in ndvi_cols[:2]:  # Plot first 2 NDVI columns
                axes[0, 1].plot(crop_yield_df[date_col], crop_yield_df[col], 'o-', label=col)
            axes[0, 1].set_title('NDVI Over Time')
            axes[0, 1].set_xlabel('Date')
            axes[0, 1].set_ylabel('NDVI')
            axes[0, 1].legend()
            axes[0, 1].tick_params(axis='x', rotation=45)

        # Plot sensor data over time if available
        sensor_cols = [col for col in crop_yield_df.columns if any(x in col.lower() for x in ['temp', 'humid', 'moist', 'rain'])]
        if sensor_cols:
            for col in sensor_cols[:3]:  # Plot first 3 sensor columns
                axes[1, 0].plot(crop_yield_df[date_col], crop_yield_df[col], 'o-', label=col)
            axes[1, 0].set_title('Sensor Data Over Time')
            axes[1, 0].set_xlabel('Date')
            axes[1, 0].set_ylabel('Value')
            axes[1, 0].legend()
            axes[1, 0].tick_params(axis='x', rotation=45)

        # Plot correlation heatmap
        numeric_cols = crop_yield_df.select_dtypes(include=[np.number]).columns.tolist()
        if len(numeric_cols) > 1:
            corr_matrix = crop_yield_df[numeric_cols].corr()
            im = axes[1, 1].imshow(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1)
            axes[1, 1].set_title('Feature Correlation Matrix')
            axes[1, 1].set_xticks(range(len(numeric_cols)))
            axes[1, 1].set_yticks(range(len(numeric_cols)))
            axes[1, 1].set_xticklabels(numeric_cols, rotation=45, ha='right')
            axes[1, 1].set_yticklabels(numeric_cols)
            plt.colorbar(im, ax=axes[1, 1])

        plt.tight_layout()
        plt.show()

        return crop_yield_df
    else:
        print("No date column found for temporal analysis")
        return crop_yield_df

def build_lstm_temporal_model(input_shape, output_units=1):
    """Build an LSTM model for temporal prediction"""
    model = Sequential([
        LSTM(64, return_sequences=True, input_shape=input_shape),
        Dropout(0.3),
        LSTM(32, return_sequences=False),
        Dropout(0.3),
        Dense(16, activation='relu'),
        Dense(output_units)
    ])

    model.compile(optimizer='adam', loss='mse', metrics=['mae', 'mse'])
    return model

# 5. Enhanced Data Processing and Model Building
def prepare_training_data(hyperspectral_data, ground_truth, crop_yield_df=None):
    """Prepare data for model training"""
    # Flatten hyperspectral data
    height, width, bands = hyperspectral_data.shape
    X_flat = hyperspectral_data.reshape(-1, bands)
    y_flat = ground_truth.reshape(-1)

    # Remove background pixels (assuming 0 is background)
    valid_indices = y_flat > 0
    X_valid = X_flat[valid_indices]
    y_valid = y_flat[valid_indices]

    print(f"Valid samples for training: {X_valid.shape[0]}")

    # Standardize features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_valid)

    # Split data
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y_valid, test_size=0.2, random_state=42, stratify=y_valid
    )

    return X_train, X_test, y_train, y_test, scaler

def build_cnn_model(input_shape, num_classes):
    """Build a CNN model for hyperspectral image classification"""
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
        MaxPooling2D((2, 2)),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Conv2D(64, (3, 3), activation='relu'),
        Flatten(),
        Dense(64, activation='relu'),
        Dropout(0.5),
        Dense(num_classes, activation='softmax')
    ])

    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    return model

def build_multimodal_model(hyperspectral_shape, sensor_shape, num_classes):
    """Build a multimodal model that combines hyperspectral and sensor data"""
    # Hyperspectral input branch
    hyperspectral_input = Input(shape=hyperspectral_shape)
    x = Conv2D(32, (3, 3), activation='relu')(hyperspectral_input)
    x = MaxPooling2D((2, 2))(x)
    x = Conv2D(64, (3, 3), activation='relu')(x)
    x = MaxPooling2D((2, 2))(x)
    x = Flatten()(x)
    x = Model(inputs=hyperspectral_input, outputs=x)

    # Sensor data input branch
    sensor_input = Input(shape=sensor_shape)
    y = Dense(16, activation='relu')(sensor_input)
    y = Model(inputs=sensor_input, outputs=y)

    # Combined model
    combined = concatenate([x.output, y.output])

    z = Dense(64, activation='relu')(combined)
    z = Dropout(0.5)(z)
    z = Dense(32, activation='relu')(z)
    z = Dense(num_classes, activation='softmax')(z)

    model = Model(inputs=[x.input, y.input], outputs=z)

    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    return model

# 6. Enhanced Dashboard and Visualization
def create_dashboard(hyperspectral_data, ground_truth, vegetation_indices, predictions=None, model=None, X_test=None, y_test=None):
    """Create an enhanced dashboard visualization"""
    ndvi, evi, ndwi, msavi2 = vegetation_indices

    fig, axes = plt.subplots(2, 3, figsize=(18, 12))

    # Original image (using a false color composite)
    axes[0, 0].imshow(hyperspectral_data[:, :, [30, 20, 10]])  # R: Red, G: Green, B: Blue
    axes[0, 0].set_title('False Color Composite')
    axes[0, 0].axis('off')

    # Ground truth
    axes[0, 1].imshow(ground_truth, cmap='tab20')
    axes[0, 1].set_title('Ground Truth')
    axes[0, 1].axis('off')

    # NDVI
    im = axes[0, 2].imshow(ndvi, cmap='RdYlGn', vmin=-1, vmax=1)
    axes[0, 2].set_title('NDVI')
    axes[0, 2].axis('off')
    plt.colorbar(im, ax=axes[0, 2])

    # EVI
    im = axes[1, 0].imshow(evi, cmap='RdYlGn', vmin=-1, vmax=1)
    axes[1, 0].set_title('EVI')
    axes[1, 0].axis('off')
    plt.colorbar(im, ax=axes[1, 0])

    # NDWI
    im = axes[1, 1].imshow(ndwi, cmap='Blues', vmin=-1, vmax=1)
    axes[1, 1].set_title('NDWI')
    axes[1, 1].axis('off')
    plt.colorbar(im, ax=axes[1, 1])

    # Confusion matrix or predictions
    if predictions is not None:
        axes[1, 2].imshow(predictions, cmap='tab20')
        axes[1, 2].set_title('Predictions')
        axes[1, 2].axis('off')
    elif model is not None and X_test is not None and y_test is not None:
        # Show confusion matrix
        y_pred = model.predict(X_test)
        y_pred_classes = np.argmax(y_pred, axis=1) if len(y_pred.shape) > 1 else y_pred
        y_true_classes = np.argmax(y_test, axis=1) if len(y_test.shape) > 1 else y_test

        cm = confusion_matrix(y_true_classes, y_pred_classes)
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[1, 2])
        axes[1, 2].set_title('Confusion Matrix')
        axes[1, 2].set_xlabel('Predicted')
        axes[1, 2].set_ylabel('True')
    else:
        axes[1, 2].axis('off')

    plt.tight_layout()
    plt.show()

# 7. Enhanced Yield Prediction
## MODIFICATION START: Added baseline model for benchmarking ##
def enhanced_yield_prediction(crop_yield_df):
    """Enhanced yield prediction with multiple models and metrics, including a baseline."""
    if crop_yield_df is None:
        return None, None, None

    # Select numeric columns only
    numeric_cols = crop_yield_df.select_dtypes(include=[np.number]).columns.tolist()
    if 'yield' not in numeric_cols:
        print("No yield column found for prediction")
        return None, None, None

    # Prepare features and target
    X_yield = crop_yield_df[numeric_cols].drop('yield', axis=1)
    y_yield = crop_yield_df['yield']

    # Handle missing values
    X_yield = X_yield.fillna(X_yield.mean())

    # Split data
    X_train_y, X_test_y, y_train_y, y_test_y = train_test_split(
        X_yield, y_yield, test_size=0.2, random_state=42
    )

    # Try multiple models, including a simple baseline for comparison
    models = {
        'Random Forest': RandomForestRegressor(n_estimators=200, random_state=42),
        'XGBoost': None,  # Would need xgboost installed
        'Baseline (Mean)': DummyRegressor(strategy='mean') # This model always predicts the mean of the training data
    }

    # Try to import XGBoost if available
    try:
        from xgboost import XGBRegressor
        models['XGBoost'] = XGBRegressor(n_estimators=200, random_state=42)
    except ImportError:
        print("XGBoost not available, skipping")
        del models['XGBoost']

    results = {}

    for name, model in models.items():
        if model is not None:
            print(f"Training {name} model...")
            model.fit(X_train_y, y_train_y)

            # Make predictions
            y_pred_y = model.predict(X_test_y)

            # Calculate metrics
            mse = mean_squared_error(y_test_y, y_pred_y)
            rmse = np.sqrt(mse)
            mae = mean_absolute_error(y_test_y, y_pred_y)
            r2 = r2_score(y_test_y, y_pred_y)

            results[name] = {
                'model': model,
                'rmse': rmse,
                'mae': mae,
                'r2': r2,
                'predictions': y_pred_y
            }

            print(f"{name} - RMSE: {rmse:.4f}, MAE: {mae:.4f}, R²: {r2:.4f}")

    print("\n--- Performance Commentary ---")
    if 'Baseline (Mean)' in results:
        baseline_r2 = results['Baseline (Mean)']['r2']
        print(f"The Baseline model achieved an R² score of {baseline_r2:.4f}. This is our benchmark.")
        print("An R² of 0.0 means the model is no better than simply guessing the average yield.")
        for name, result in results.items():
            if name != 'Baseline (Mean)':
                if result['r2'] > baseline_r2:
                    print(f"-> The '{name}' model (R²: {result['r2']:.4f}) shows a significant improvement over the baseline, indicating it has learned valuable patterns from the data.")
                else:
                    print(f"-> The '{name}' model (R²: {result['r2']:.4f}) did NOT perform better than the baseline. Its predictions are not reliable.")
    print("----------------------------\n")

    # Plot results
    if results:
        fig, axes = plt.subplots(1, 2, figsize=(15, 6))

        # Plot actual vs predicted
        for name, result in results.items():
            axes[0].scatter(y_test_y, result['predictions'], alpha=0.5, label=name)

        axes[0].plot([y_test_y.min(), y_test_y.max()], [y_test_y.min(), y_test_y.max()], 'k--', lw=2)
        axes[0].set_xlabel('Actual Yield')
        axes[0].set_ylabel('Predicted Yield')
        axes[0].set_title('Actual vs Predicted Yield')
        axes[0].legend()

        # Plot metrics comparison
        metrics_df = pd.DataFrame({name: [result['rmse'], result['mae'], result['r2']]
                                   for name, result in results.items()},
                                  index=['RMSE', 'MAE', 'R²'])
        metrics_df.plot(kind='bar', ax=axes[1])
        axes[1].set_title('Model Performance Comparison')
        axes[1].set_ylabel('Score')
        axes[1].tick_params(axis='x', rotation=0)

        plt.tight_layout()
        plt.show()

    return results, X_train_y, X_test_y
## MODIFICATION END ##

# 8. Chatbot Integration (Simplified)
class AgricultureChatbot:
    """Simple chatbot for agricultural queries"""
    def __init__(self):
        self.responses = {
            "hello": "Hello! I'm your agriculture assistant. How can I help with crop monitoring today?",
            "hi": "Hi there! I can help you with crop health, soil conditions, and pest risks.",
            "ndvi": "NDVI (Normalized Difference Vegetation Index) measures vegetation health. Values range from -1 to 1, with higher values indicating healthier vegetation.",
            "pest": "I can help identify pests from images. Upload an insect image for analysis.",
            "yield": "Based on current data, your predicted yield is calculated using vegetation indices and sensor data.",
            "soil": "Soil health can be monitored using various indices and sensor data like moisture content.",
            "help": "I can help with: NDVI calculation, pest identification, yield prediction, and soil health monitoring.",
            "default": "I'm not sure how to help with that. Try asking about NDVI, pests, yield, or soil health."
        }

    def get_response(self, query):
        query = query.lower()
        for key in self.responses:
            if key in query:
                return self.responses[key]
        return self.responses["default"]

# 9. Enhanced Streamlit App Preparation
def generate_streamlit_app():
    """Generate enhanced code for a Streamlit dashboard"""
    streamlit_code = '''
import streamlit as st
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from PIL import Image
import io
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Set up the Streamlit app
st.set_page_config(page_title="Crop Health Monitoring", layout="wide")
st.title("AI-Powered Crop Health Monitoring System")

# Sidebar for navigation
st.sidebar.title("Navigation")
options = st.sidebar.selectbox("Select a page:",
                                ["Dashboard", "Vegetation Indices", "Pest Detection", "Yield Prediction", "Temporal Analysis", "Chatbot"])

if options == "Dashboard":
    st.header("Farm Dashboard")
    col1, col2 = st.columns(2)

    with col1:
        st.subheader("Field Overview")
        # Display map or field image
        st.image("https://via.placeholder.com/400x300?text=Field+Map", caption="Field Map")

    with col2:
        st.subheader("Key Metrics")
        st.metric("Average NDVI", "0.68", "0.12")
        st.metric("Pest Risk", "Low", "-2%")
        st.metric("Yield Forecast", "5.2 tons/acre", "0.3 tons")

    # Recent alerts
    st.subheader("Recent Alerts")
    st.warning("Northern section shows signs of water stress")
    st.info("Soil moisture optimal in southern fields")

    # NDVI trend
    st.subheader("NDVI Trend (Last 30 Days)")
    trend_data = pd.DataFrame({
        'Date': pd.date_range(start='2023-01-01', periods=30),
        'NDVI': np.random.uniform(0.6, 0.8, 30)
    })
    fig = px.line(trend_data, x='Date', y='NDVI', title='NDVI Trend Over Time')
    st.plotly_chart(fig, use_container_width=True)

elif options == "Vegetation Indices":
    st.header("Vegetation Indices Analysis")

    # Upload multispectral image
    uploaded_file = st.file_uploader("Upload multispectral image", type=["tif", "jpg", "png"])

    if uploaded_file is not None:
        # Process image and calculate indices
        image = Image.open(uploaded_file)
        st.image(image, caption="Uploaded Image", use_column_width=True)

        # Display indices (placeholder)
        col1, col2, col3, col4 = st.columns(4)
        with col1:
            st.subheader("NDVI")
            st.image("https://via.placeholder.com/200x150?text=NDVI+Map", caption="NDVI Map")
            st.metric("Average NDVI", "0.72", "0.05")

        with col2:
            st.subheader("EVI")
            st.image("https://via.placeholder.com/200x150?text=EVI+Map", caption="EVI Map")
            st.metric("Average EVI", "0.45", "0.03")

        with col3:
            st.subheader("NDWI")
            st.image("https://via.placeholder.com/200x150?text=NDWI+Map", caption="NDWI Map")
            st.metric("Average NDWI", "0.28", "-0.02")

        with col4:
            st.subheader("MSAVI2")
            st.image("https://via.placeholder.com/200x150?text=MSAVI2+Map", caption="MSAVI2 Map")
            st.metric("Average MSAVI2", "0.62", "0.04")

        # Show health status
        st.subheader("Field Health Status")
        health_data = pd.DataFrame({
            'Section': ['North', 'South', 'East', 'West'],
            'NDVI': [0.68, 0.72, 0.65, 0.71],
            'Health': ['Moderate', 'Good', 'Moderate', 'Good']
        })
        fig = px.bar(health_data, x='Section', y='NDVI', color='Health',
                     title='NDVI by Field Section')
        st.plotly_chart(fig, use_container_width=True)

elif options == "Pest Detection":
    st.header("Pest Detection")

    uploaded_file = st.file_uploader("Upload insect image", type=["jpg", "jpeg", "png"])

    if uploaded_file is not None:
        image = Image.open(uploaded_file)
        st.image(image, caption="Uploaded Insect Image", use_container_width=True)

        # Placeholder for pest detection results with confidence heatmap
        st.subheader("Detection Results")
        col1, col2 = st.columns(2)

        with col1:
            st.write("**Identified Pest:** Aphids")
            st.write("**Confidence:** 87%")
            st.write("**Risk Level:** High")
            st.write("**Recommended Action:**")
            st.write("- Apply neem oil solution")
            st.write("- Introduce ladybugs as natural predators")
            st.write("- Monitor affected area daily")

        with col2:
            # Confidence heatmap (placeholder)
            st.write("**Detection Confidence:**")
            pest_confidences = {
                'Aphids': 0.87,
                'Whiteflies': 0.08,
                'Spider Mites': 0.03,
                'Other': 0.02
            }
            fig = px.bar(x=list(pest_confidences.keys()), y=list(pest_confidences.values()),
                         labels={'x': 'Pest Type', 'y': 'Confidence'},
                         title='Pest Detection Confidence')
            st.plotly_chart(fig, use_container_width=True)

        # Show historical pest occurrences
        st.subheader("Historical Pest Occurrences")
        history_data = pd.DataFrame({
            'Date': pd.date_range(start='2023-01-01', periods=12, freq='M'),
            'Aphids': np.random.randint(0, 10, 12),
            'Whiteflies': np.random.randint(0, 5, 12),
            'Spider Mites': np.random.randint(0, 3, 12)
        })
        fig = px.line(history_data, x='Date', y=['Aphids', 'Whiteflies', 'Spider Mites'],
                      title='Pest Occurrences Over Time')
        st.plotly_chart(fig, use_container_width=True)

elif options == "Yield Prediction":
    st.header("Yield Prediction")

    # Input parameters
    col1, col2 = st.columns(2)

    with col1:
        ndvi = st.slider("Average NDVI", 0.0, 1.0, 0.7)
        soil_moisture = st.slider("Soil Moisture (%)", 0.0, 100.0, 65.0)
        rainfall = st.slider("Rainfall (mm)", 0.0, 500.0, 120.0)

    with col2:
        temperature = st.slider("Temperature (°C)", 10.0, 40.0, 25.0)
        nitrogen = st.slider("Nitrogen Level (kg/ha)", 0.0, 200.0, 100.0)
        phosphorus = st.slider("Phosphorus Level (kg/ha)", 0.0, 100.0, 50.0)

    # Predict yield
    if st.button("Predict Yield"):
        # Enhanced prediction formula
        predicted_yield = (2.5 + (ndvi * 3.2) + (soil_moisture * 0.02) +
                           (rainfall * 0.005) - (temperature * 0.03) +
                           (nitrogen * 0.01) + (phosphorus * 0.015))

        st.success(f"Predicted Yield: {predicted_yield:.2f} tons/acre")

        # Show historical comparison
        st.subheader("Historical Comparison")
        historical_data = pd.DataFrame({
            'Year': [2018, 2019, 2020, 2021, 2022, 2023],
            'Yield': [4.2, 4.5, 4.8, 5.1, 4.9, predicted_yield]
        })
        fig = px.line(historical_data, x='Year', y='Yield',
                      title='Historical Yield Comparison', markers=True)
        st.plotly_chart(fig, use_container_width=True)

        # Show feature importance
        st.subheader("Feature Importance")
        importance_data = pd.DataFrame({
            'Feature': ['NDVI', 'Soil Moisture', 'Rainfall', 'Temperature', 'Nitrogen', 'Phosphorus'],
            'Importance': [0.35, 0.25, 0.15, 0.10, 0.08, 0.07]
        })
        fig = px.bar(importance_data, x='Importance', y='Feature', orientation='h',
                     title='Feature Importance for Yield Prediction')
        st.plotly_chart(fig, use_container_width=True)

elif options == "Temporal Analysis":
    st.header("Temporal Analysis")

    # Generate sample temporal data
    dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
    temporal_data = pd.DataFrame({
        'Date': dates,
        'NDVI': 0.6 + 0.2 * np.sin(np.arange(len(dates)) / 30) + np.random.normal(0, 0.05, len(dates)),
        'Temperature': 20 + 10 * np.sin(np.arange(len(dates)) / 30) + np.random.normal(0, 2, len(dates)),
        'Rainfall': np.random.exponential(2, len(dates)),
        'Soil Moisture': 50 + 20 * np.sin(np.arange(len(dates)) / 30) + np.random.normal(0, 5, len(dates))
    })

    # NDVI trend
    st.subheader("NDVI Trend Over Time")
    fig = px.line(temporal_data, x='Date', y='NDVI', title='NDVI Trend')
    st.plotly_chart(fig, use_container_width=True)

    # Multiple metrics
    st.subheader("Multiple Metrics Over Time")
    fig = make_subplots(rows=2, cols=2, subplot_titles=('NDVI', 'Temperature', 'Rainfall', 'Soil Moisture'))

    fig.add_trace(go.Scatter(x=temporal_data['Date'], y=temporal_data['NDVI'], name='NDVI'), row=1, col=1)
    fig.add_trace(go.Scatter(x=temporal_data['Date'], y=temporal_data['Temperature'], name='Temperature'), row=1, col=2)
    fig.add_trace(go.Bar(x=temporal_data['Date'], y=temporal_data['Rainfall'], name='Rainfall'), row=2, col=1)
    fig.add_trace(go.Scatter(x=temporal_data['Date'], y=temporal_data['Soil Moisture'], name='Soil Moisture'), row=2, col=2)

    fig.update_layout(height=600, showlegend=False)
    st.plotly_chart(fig, use_container_width=True)

    # Correlation analysis
    st.subheader("Correlation Analysis")
    corr_matrix = temporal_data.drop('Date', axis=1).corr()
    fig = px.imshow(corr_matrix, text_auto=True, aspect="auto",
                    title='Correlation Between Metrics')
    st.plotly_chart(fig, use_container_width=True)

elif options == "Chatbot":
    st.header("Agriculture Assistant Chatbot")

    # Initialize chat history
    if "messages" not in st.session_state:
        st.session_state.messages = []

    # Display chat messages from history on app rerun
    for message in st.session_state.messages:
        with st.chat_message(message["role"]):
            st.markdown(message["content"])

    # React to user input
    if prompt := st.chat_input("Ask about crop health, pests, or yield"):
        # Display user message in chat message container
        st.chat_message("user").markdown(prompt)
        # Add user message to chat history
        st.session_state.messages.append({"role": "user", "content": prompt})

        # Enhanced response logic
        response = "I can help with NDVI, pest identification, yield prediction, and soil health monitoring. Ask me about any of these topics!"
        if "ndvi" in prompt.lower():
            response = "NDVI (Normalized Difference Vegetation Index) measures vegetation health. Values range from -1 to 1, with higher values indicating healthier vegetation. Your current field NDVI is 0.72, which indicates good health."
        elif "pest" in prompt.lower():
            response = "I can help identify pests from images. Common agricultural pests include aphids, caterpillars, and beetles. Upload an image for identification. Based on recent data, your pest risk is currently low."
        elif "yield" in prompt.lower():
            response = "Yield prediction uses vegetation indices, weather data, and soil conditions to forecast crop production. Current models predict a yield of 5.2 tons/acre for your field, which is above average for this season."
        elif "soil" in prompt.lower():
            response = "Soil health can be monitored using various indices and sensor data. Your current soil moisture is at 65%, which is optimal for most crops. Soil temperature is 22°C and nitrogen levels are adequate."
        elif "recommend" in prompt.lower():
            response = "Based on current conditions, I recommend: 1) Monitor NDVI weekly, 2) Check for pests in the northern section, 3) Consider irrigation if rainfall remains below average."

        # Display assistant response in chat message container
        with st.chat_message("assistant"):
            st.markdown(response)
        # Add assistant response to chat history
        st.session_state.messages.append({"role": "assistant", "content": response})

# Footer
st.sidebar.markdown("---")
st.sidebar.info("AI-Powered Crop Health Monitoring System | © 2023")
'''

    # Save the Streamlit app code to a file
    with open(DATASET_PATH / 'crop_health_dashboard.py', 'w') as f:
        f.write(streamlit_code)

    print("Enhanced Streamlit app code generated at: crop_health_dashboard.py")

# 10. Main execution
def main():
    print("Starting AI-powered Crop Health Monitoring System")

    # Load datasets
    print("\n1. Loading datasets...")
    hyperspectral_data, ground_truth = load_salinas_data()
    crop_yield_df = load_crop_yield_data()
    insect_images, insect_labels, insect_label_map = load_farm_insects()
    lc08_files, lc08_data, lc08_transform, lc08_crs = load_lc08_data()
    sentinel_bands, safe_folder = load_safe_data()

    if hyperspectral_data is None:
        print("No hyperspectral data found. Using alternative data sources.")

    # Calculate vegetation indices
    print("\n2. Calculating vegetation indices...")
    if hyperspectral_data is not None:
        ndvi, evi, ndwi, msavi2 = extract_vegetation_indices(hyperspectral_data)
        vegetation_indices = (ndvi, evi, ndwi, msavi2)
    elif sentinel_bands:
        # Use Sentinel-2 indices if available
        sentinel_indices = calculate_sentinel_indices(sentinel_bands)
        vegetation_indices = (
            sentinel_indices.get('ndvi', None),
            sentinel_indices.get('evi', None),
            sentinel_indices.get('ndwi', None),
            sentinel_indices.get('msavi2', None)
        )
        print("Calculated vegetation indices from Sentinel-2 data")
    else:
        # Use Landsat indices if available
        if lc08_files:
            landsat_indices = extract_landsat_indices(lc08_files)
            if landsat_indices:
                vegetation_indices = (
                    landsat_indices.get('ndvi', None),
                    landsat_indices.get('evi', None),
                    landsat_indices.get('ndwi', None),
                    None
                )
                print("Calculated vegetation indices from Landsat data")
            else:
                vegetation_indices = (None, None, None, None)
        else:
            vegetation_indices = (None, None, None, None)

    # Create initial dashboard if we have data
    if hyperspectral_data is not None and ground_truth is not None:
        print("\n3. Creating initial dashboard...")
        create_dashboard(hyperspectral_data, ground_truth, vegetation_indices)


    if insect_images and insect_labels:
        print("\n4. Training and Comparing Pest Detection Models...")
        X_insect, y_insect = prepare_insect_data_augmented(insect_images, insect_labels, img_size=(224, 224))

        # Calculate class weights for imbalanced data
        class_weights = calculate_class_weights(y_insect)
        print(f"Class weights: {class_weights}")

        # Split data
        X_train_i, X_test_i, y_train_i, y_test_i = train_test_split(
            X_insect, y_insect, test_size=0.2, random_state=42, stratify=y_insect
        )

        # --- Model Comparison Loop ---
        model_types_to_test = ['efficientnet', 'resnet', 'mobilenet']
        pest_model_results = {}

        for model_type in model_types_to_test:
            print(f"\n--- Training Pest Detection Model: {model_type.upper()} ---")

            # Build and train model with transfer learning
            pest_model, preprocess_fn = build_pest_detection_model_transfer(
                X_train_i[0].shape, len(insect_label_map), model_type=model_type
            )

            # Set up callbacks
            callbacks = [
                EarlyStopping(patience=10, restore_best_weights=True),
                ReduceLROnPlateau(factor=0.2, patience=5)
            ]

            # Train model
            history = pest_model.fit(
                X_train_i, y_train_i,
                epochs=50,  # Train for longer
                validation_split=0.2,
                class_weight=class_weights,
                callbacks=callbacks,
                verbose=1
            )

            # Evaluate model
            test_loss, test_acc = pest_model.evaluate(X_test_i, y_test_i, verbose=0)
            print(f"{model_type.upper()} - Test Loss: {test_loss:.4f}, Test Accuracy: {test_acc:.4f}")
            pest_model_results[model_type] = test_acc

            # Plot training history
            plt.figure(figsize=(12, 4))
            plt.suptitle(f'Training History for {model_type.upper()}', fontsize=16)
            plt.subplot(1, 2, 1)
            plt.plot(history.history['accuracy'], label='Training Accuracy')
            plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
            plt.title('Model Accuracy')
            plt.xlabel('Epoch')
            plt.ylabel('Accuracy')
            plt.legend()

            plt.subplot(1, 2, 2)
            plt.plot(history.history['loss'], label='Training Loss')
            plt.plot(history.history['val_loss'], label='Validation Loss')
            plt.title('Model Loss')
            plt.xlabel('Epoch')
            plt.ylabel('Loss')
            plt.legend()

            plt.tight_layout(rect=[0, 0.03, 1, 0.95])
            plt.show()

        # --- Print Final Model Comparison Summary ---
        print("\n--- Pest Detection Model Comparison ---")
        best_model = max(pest_model_results, key=pest_model_results.get)
        for model_type, acc in pest_model_results.items():
            print(f"  - {model_type.upper()}: Test Accuracy = {acc:.4f}")
        print(f"\n🏆 Best performing model: {best_model.upper()} with an accuracy of {pest_model_results[best_model]:.4f}")
        print("-------------------------------------\n")


    if crop_yield_df is not None:
        print("\n5. Analyzing temporal trends...")
        crop_yield_df = analyze_temporal_trends(crop_yield_df)

    # Prepare training data and train models if we have hyperspectral data
    if hyperspectral_data is not None and ground_truth is not None:
        print("\n6. Training crop classification model...")
        X_train, X_test, y_train, y_test, scaler = prepare_training_data(
            hyperspectral_data, ground_truth, crop_yield_df
        )

        # Train a simple Random Forest classifier
        rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
        rf_model.fit(X_train, y_train)

        # Evaluate the model
        y_pred = rf_model.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)
        print(f"Random Forest Accuracy: {accuracy:.4f}")

        # Show detailed classification report
        print("\nClassification Report:")
        print(classification_report(y_test, y_pred))

        # Make predictions on the entire image
        print("\n7. Making predictions on entire image...")
        height, width, bands = hyperspectral_data.shape
        X_full_flat = hyperspectral_data.reshape(-1, bands)

        # Standardize and predict
        X_full_scaled = scaler.transform(X_full_flat)
        full_pred_flat = rf_model.predict(X_full_scaled)
        full_pred = full_pred_flat.reshape(height, width)

        # Create enhanced dashboard with predictions and confusion matrix
        create_dashboard(hyperspectral_data, ground_truth, vegetation_indices, full_pred, rf_model, X_test, y_test)

    # Enhanced yield prediction with multiple models
    if crop_yield_df is not None:
        print("\n8. Building enhanced yield prediction models...")
        yield_results, X_train_y, X_test_y = enhanced_yield_prediction(crop_yield_df)

    # Generate enhanced Streamlit app code
    print("\n9. Generating enhanced Streamlit dashboard code...")
    generate_streamlit_app()

    # Initialize chatbot
    print("\n10. Initializing agriculture chatbot...")
    chatbot = AgricultureChatbot()

    # Test chatbot
    test_queries = ["hello", "what is ndvi?", "pest detection", "yield prediction", "soil health"]
    print("\nChatbot test:")
    for query in test_queries:
        response = chatbot.get_response(query)
        print(f"Q: {query}")
        print(f"A: {response}\n")

    print("\n11. System ready for monitoring crop health, soil condition, and pest risks!")
    print("To run the web dashboard, execute: streamlit run crop_health_dashboard.py")

# Run the main function
if __name__ == "__main__":
    main()
```
